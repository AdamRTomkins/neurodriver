# Artificial LPU Integration Demo

This notebook illustrates how to define local processing unit (LPU) models based upon Neurokernel and integrate them into a single model using inter-LPU connectivity patterns.

#### Configuration

An LPU comprises two distinct populations of neurons [[1]](#chang_three-dimensional_2011): *local* neurons may only connect to other neurons in the LPU, while *projection* neurons may connect both to local neurons and neurons in other LPUs. *Input* neurons can be understood to correspond to sensory neurons in biological LPUs. Both local and projection neurons may receive input from input neurons (via the synapses depicted in light yellow), neurons within the LPU (via the synapses depicted in light pink), or neurons in other LPUs (via the synapses depicted in light green). The general structure of an LPU is shown below:

```python
%cd -q ~/neurokernel/notebooks/
import IPython.display
IPython.display.Image(filename='files/artificial-lpu.png')
```

The artificial LPUs employed in this example are randomly generated: the number of neurons in each populations is randomly selected between 30 to 40. The input neurons and ouput neurons are modeled as leaky integrate-and-fire (LIF) neurons, and each of the local neurons is modeled as either an LIF neuron or a graded potential neuron that employs the Morris-Lecar model but outputs the model's membrane potential rather than its spikes. Synaptic currents are modeled with alpha functions. Synapses between each pair of neuron populations (i.e., input neurons, projection neurons, and local neurons) are also created randomly; roughly half of the total number of pairs of neurons are connected. To generate the LPUs and input signals used in this demo, we run the following script:

```python
%cd -q ~/neurokernel/examples/integration
%run gen_artificial_lpu.py -s 0 artificial_lpu_0.gexf.gz artificial_input_0.h5
%run gen_artificial_lpu.py -s 1 artificial_lpu_1.gexf.gz artificial_input_1.h5
```

#### Executing the Model

Once the configuration and the input stimulus are ready, we execute the entire model both with and without connections defined between the LPUs:
